```python
#CREATE BAR CHARTS USING THE OBJECT ORIENTED APPROACH
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
fig,ax=plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
fig,ax=plt.subplots()
ax.barh(y_axis,x_axis)
```

```python
fig,ax=plt.subplots()
ax.barh(x_axis,y_axis,color='cyan',label='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('Pyber fare by month')
plt.gca().invert_yaxis()
plt.legend()

```

```python
#create scatter plots and bubble charts
```

```python
plt.plot(x_axis,y_axis,'o')
```

```python
plt.scatter(y_axis,x_axis,color='red',label='Chicago')
plt.xlabel('Fare')
plt.ylabel('Date')
plt.gca().invert_yaxis()
plt.legend()
```

```python
#create a bubble chart using MTALAB method
```

```python
plt.scatter(x_axis,y_axis,s=y_axis)
```

```python
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
#new_list=[expression for item in list if conditional]
```

```python
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
#create a scatter plot using the objected oriented interface
```

```python
fig,ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
fig,ax=plt.subplots()
ax.scatter(x_axis,y_axis,s=y_axis)
```

```python
fig,ax=plt.subplots()
ax.scatter(y_axis,x_axis,s=[i*5 for i in y_axis],alpha=.2,edgecolor='black',linewidth=2,label='Boston')
plt.title('title')
plt.xlabel('fare')
plt.ylabel('date')
plt.xlim(0,50)
plt.gca().invert_xaxis()
plt.legend()

```

```python
#Create Pie charts#
```

```python
plt.pie(y_axis,labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,explode=explode_values,labels=x_axis,autopct='%.1f%%')
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,explode=explode_values,colors=colors,labels=x_axis,autopct='%.1f%%')
```

```python
#creating a pie chart USING OBJECT ORIENTED INTERFACE
```

```python
fig, ax=plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
fig, ax=plt.subplots()
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode=[0,0,.2,0,0,0,.2,0,0,0,0,0]
plt.subplots(figsize=(8,8))
plt.pie(y_axis,labels=x_axis,explode=explode,shadow=True,autopct='%.1f%%',counterclock=False,startangle=True)
plt.show()
```
